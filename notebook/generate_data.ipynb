```python
import sqlite3
import random
from datetime import datetime, timedelta

conn = conn = sqlite3.connect(
    r"C:\Users\dipas\product_analytics_project\data\product_analytics.db"
)

cursor = conn.cursor()

cursor.execute("DELETE FROM events")
cursor.execute("DELETE FROM users")
cursor.execute("DELETE FROM subscriptions")
conn.commit()

countries = ["India", "US", "UK"]
devices = ["Web", "Mobile"]

start_date = datetime(2025, 1, 1)

user_count = 500
event_id = 1

for user_id in range(1, user_count + 1):
    signup_date = start_date + timedelta(days=random.randint(0, 30))
    country = random.choice(countries)
    device = random.choice(devices)

    
    cursor.execute(
        "INSERT INTO users VALUES (?, ?, ?, ?)",
        (user_id, signup_date.date(), country, device)
    )

   
    cursor.execute(
        "INSERT INTO events VALUES (?, ?, ?, ?)",
        (event_id, user_id, "signup", signup_date)
    )
    event_id += 1

  
    if random.random() < 0.8:
        login_time = signup_date + timedelta(hours=random.randint(1, 48))
        cursor.execute(
            "INSERT INTO events VALUES (?, ?, ?, ?)",
            (event_id, user_id, "login", login_time)
        )
        event_id += 1

        
        if random.random() < 0.6:
            feature_time = login_time + timedelta(days=random.randint(0, 7))
            cursor.execute(
                "INSERT INTO events VALUES (?, ?, ?, ?)",
                (event_id, user_id, "feature_use", feature_time)
            )
            event_id += 1

       
            if random.random() < 0.3:
                purchase_time = feature_time + timedelta(days=random.randint(1, 5))
                cursor.execute(
                    "INSERT INTO events VALUES (?, ?, ?, ?)",
                    (event_id, user_id, "purchase", purchase_time)
                )
                event_id += 1

conn.commit()
conn.close()

print("Data generation completed.")
```
